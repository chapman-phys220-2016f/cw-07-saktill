# Problem 1: Exercise C.1)
Solve the ODE problem
$$u' = 2u − 1, u(0) = 2, t ∈ [0, 6]$$
using the Forward Euler method. Choose $∆t = 0.25$. Plot the numerical
solution together with the exact solution $$u(t) = \frac{1}{2} + 2e^{2t}$$
.

```python
l = nlo.solve_ODE(2,6,24)#initial condition 2, end time 6, samplepoints 24. timestep = 0.24 (0-6)/(24)
u_approx = l[0]
t = l[1]
u_exact = lambda t: 0.5 + 2*np.exp(2*t)
u = u_exact(t)
plt.plot(t,u_approx,'r-')
plt.plot(t, u,'b-')
plt.xlabel('t')
plt.ylabel('u')
plt.title('Euler method vs. Exact solution')
plt.show()
```

With the specified $∆t$ from the problem, we get a pretty bad approximation (red) to the exact curve (blue). But if we increase our sample points to $n=1000$ and decrease our time-step to $∆t = 0.006$ we see a better approximation as shown below.

```python
l1 = nlo.solve_ODE(2,6,1000)
u_approx1 = l1[0]
t1 = l1[1]
u_exact1 = lambda t: 0.5 + 2*np.exp(2*t)
u1 = u_exact1(t1)
plt.plot(t1,u_approx1,'r-')
plt.plot(t1, u1,'b-')
plt.xlabel('t')
plt.ylabel('u')
plt.title('Euler method vs. Exact solution')
plt.show()
```

## Problem 2: Exercise C2
Solve the non-linear ODE problem: $$u' = u^q,  \quad  u(0) = 1, \quad   t \in [0,T]$$
Run for $q = 2,3$ and for $\Delta t = .1,.01$
Plot both the exact solution and the numerical approximation via Euler's Forward method.

The exact solution was given, so to implement the Forward Euler approximation, I created a difference equation for each element of a mesh representation of $u$. Here are the four plots:

```python
lExact = node.solve(2,.1)
lApprox = node.numSolve(2,.1)
plt.plot(lExact[0],lExact[1],'b-')
plt.plot(lApprox[0],lApprox[1],'r-')
plt.legend(['Exact','Approx'])
plt.title("u' = u^2,  dt=.1")
plt.show()
```

```python
lExact = node.solve(2,.01)
lApprox = node.numSolve(2,.01)
plt.plot(lExact[0],lExact[1],'b-')
plt.plot(lApprox[0],lApprox[1],'r-')
plt.legend(['Exact','Approx'])
plt.title("u' = u^2,  dt=.01")
plt.show()
```

```python
lExact = node.solve(3,.1)
lApprox = node.numSolve(3,.1)
plt.plot(lExact[0],lExact[1],'b-')
plt.plot(lApprox[0],lApprox[1],'r-')
plt.legend(['Exact','Approx'])
plt.title("u' = u^3,  dt=.1")
plt.show()
```

```python
lExact = node.solve(3,.01)
lApprox = node.numSolve(3,.01)
plt.plot(lExact[0],lExact[1],'b-')
plt.plot(lApprox[0],lApprox[1],'r-')
plt.legend(['Exact','Approx'])
plt.title("u' = u^3,  dt=.01")
plt.show()
```

As expected, a smaller step size results in a more accurate plot, but additionally, the inaccuracy of $u' = u^3$ is not much more than $u' = u^2$. And also as expected, the error builds on itself with each step, so the plots diverge more for higher $t$.
